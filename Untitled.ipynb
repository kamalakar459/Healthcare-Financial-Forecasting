```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cvs = yf.download("CVS", start="2019-01-01", end="2024-12-31")
cvs.head()

```

Data Exploration

We begin by downloading 5 years of CVS Health (CVS) stock data and exploring price movements,  
daily returns, and trading activity. This helps us understand historical patterns and volatility  
before applying predictive models.


```python
plt.figure(figsize=(12,6))
plt.plot(cvs['Close'], label="CVS Closing Price", color='blue')
plt.title("CVS Health Closing Price (2019â€“2024)")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

```python
cvs.info()
```

```python
cvs["MA50"] = cvs["Close"].rolling(50).mean()
cvs["MA200"] = cvs["Close"].rolling(200).mean()
plt.figure(figsize=(12,6))
plt.plot(cvs["Close"], label="Closing Price", color="blue")
plt.plot(cvs["MA50"], label="50-Day MA", color="red")
plt.plot(cvs["MA200"], label="200-Day MA", color="green")
plt.title("CVS Stock with 50 & 200 Day Moving Averages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(cvs.index, cvs["Volume"], color="purple")
plt.title("CVS Trading Volume (2019â€“2024)")
plt.xlabel("Date")
plt.ylabel("Shares Traded")
plt.show()
```

```python
cvs["Daily Return"] = cvs["Close"].pct_change()
plt.figure(figsize=(12,6))
plt.plot(cvs["Daily Return"], color="orange")
plt.title("CVS Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily % Change")
plt.show()
cvs["Daily Return"].describe()
```

```python
cvs["Volatility"] = cvs["Daily Return"].rolling(30).std()
plt.figure(figsize=(12,6))
plt.plot(cvs["Volatility"], color="red")
plt.title("CVS 30-Day Rolling Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```

```python
Forecasting with ARIMA
ARIMA (AutoRegressive Integrated Moving Average) is a widely used time-series model in finance.  
We fit the model to CVS stock prices and forecasted the next 90 business days.
```

```python
%pip install statsmodels
```

```python
from statsmodels.tsa.arima.model import ARIMA

data = cvs["Close"].dropna()
model = ARIMA(data, order=(5,1,0))
model_fit = model.fit()
forecast = model_fit.forecast(steps=90)
plt.figure(figsize=(12,6))
plt.plot(data, label="Historical Closing Price")
plt.plot(pd.date_range(data.index[-1], periods=91, freq="B")[1:], forecast, 
         label="Forecast", color="red")
plt.title("CVS Stock Price Forecast (Next 90 Days)")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```
